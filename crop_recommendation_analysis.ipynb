```python
# IMPORT LIBRARIES 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

```

```python
#DATA LOADING
df= pd.read_csv("Test Dataset.csv")

shape=df.shape
print(shape)

head= df.head()
print(head)

print(df.isnull().sum())


describe= df.describe()
print(describe)


#DATA CLEANING
df = df.drop(columns=["Unnamed: 0"])
```

```python
# DATA ANALYSIS BY VISUALIZATION 

#DISTRIBUTION OF NITROGEN 
plt.figure(figsize=(10,5))
sns.histplot(df["N"], kde=True)
plt.xlabel("Nitrogen")
plt.title("Distribution of nitrogen levels")
plt.show()


#CROP FREQUENCY 
plt.figure(figsize=(15,10))
sns.countplot(y="Crop",data=df)
plt.title("Crop Distribution")
plt.show()



# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Soil Feature Correlation Heatmap")
plt.show()


```

```python
crop_group=df.groupby("Crop").mean()
print(crop_group)

plt.figure(figsize=(15,10))
crop_group[["N", "P", "K"]].plot(kind="bar")
plt.title("Average NPK requirements by different crops")
plt.ylabel("Nutrient Level")
plt.show()
```
